```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score,classification_report
```

```python
df=pd.read_csv('pandaCSV/lifeexpectancydata.csv')
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
print(df.isna().sum())
```

```python
df.dropna(inplace=True,axis=0)
print(df.isna().sum())
```

```python
df.shape
```

```python
df.head(3)
```

```python
x=df.drop(columns=['Life expectancy','Country','Status'])
y=df[['Life expectancy']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=5)
x_train.shape
```

```python
# creating correlation matrix
correlation_matrix=x.corr()
print(correlation_matrix)
plt.figure(figsize=(15,15))

# visualizing with seaborn as a heatmap
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
lr_model = LinearRegression()
lr_model.fit(x_train,y_train)
lr_model_preds = lr_model.predict(x_test)
print(f'MSE = {mean_squared_error(y_test,lr_model_preds)}')
print(f'MAE = {mean_absolute_error(y_test,lr_model_preds)}')
print(f'R2 Score = {r2_score(y_test,lr_model_preds)}')
```
